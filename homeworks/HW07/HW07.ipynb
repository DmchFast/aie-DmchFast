### **2.3.1.** Загрузка данных и первичный анализ `S07-hw-dataset-01.csv`, `S07-hw-dataset-02.csv` и `S07-hw-dataset-04.csv`

```python
import pandas as pd
from pathlib import Path

path_files = Path("data")

datasets = ["S07-hw-dataset-01.csv", "S07-hw-dataset-02.csv", "S07-hw-dataset-04.csv",]

dataframes = {}

for f in datasets:
    print(f"-----------------Датасет: {f}-----------------")
    
    # Загрузка csv
    df = pd.read_csv(path_files / f)
    dataframes[f] = df
    
    # Первичный анализ
    print("\nhead()")
    print(display(df.head()))
    print("\ninfo()")
    df.info()
    print("\ndescribe() (числовые признаки)")
    print(display(df.describe()))
    
    # Пропуски в датасетах
    missing_abs = df.isna().sum()
    missing_rel = df.isna().mean()
    
    print("\nПропуски (абсол их кол) ->", display(missing_abs[missing_abs > 0]))
    print("\nПропуски (их доля) ->", display(missing_rel[missing_rel > 0]))
    
    # Признаки и их типы
    numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
    categorical_cols = df.select_dtypes(exclude=["number"]).columns.tolist()
    print("\nТипы признаков:")
    print(f"Числовые ({len(numeric_cols)}) -> {numeric_cols}")
    print(f"Категориальные ({len(categorical_cols)}) -> {categorical_cols}")
    
    # X и sample_id
    sample_id = df["sample_id"]
    X = df.drop(columns=["sample_id"])
    print("\nПроверка X и sample_id")
    print("X shape:", X.shape)
    print("sample_id shape:", sample_id.shape)
    print()
```

### **2.3.2.** Препроцессинг

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

preprocesor = {}

for f, df in dataframes.items():
    print(f"-----------------Препроцессинг для {f}-----------------")
    
    # Разделение на  числовые и категор признаки
    numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
    categorical_cols = df.select_dtypes(exclude=["number"]).columns.tolist()
    
    if "sample_id" in numeric_cols:
        numeric_cols.remove("sample_id")
    if "sample_id" in categorical_cols:
        categorical_cols.remove("sample_id")
    
    # Препроцессинг числовых признаков
    numeric_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')), 
        ('scaler', StandardScaler())
    ])
    
    # Препроцессинг категор признаков 
    if len(categorical_cols) > 0:
        categorical_transformer = Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])
    else:
        categorical_transformer = 'passthrough'

    # ColumnTransformer
    preprocessor = ColumnTransformer([
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
    
    preprocesor[f] = preprocessor
    print(f"Числовые признаки -> {numeric_cols}")
    print(f"Категориальные признаки -> {categorical_cols if len(categorical_cols)>0 else 'нет'}")
    print("Готово\n")

# Использование в S07-hw-dataset-01.csv
X = dataframes["S07-hw-dataset-01.csv"].drop(columns=["sample_id"])
X_transformed = preprocesor["S07-hw-dataset-01.csv"].fit_transform(X)
print("После трансформации в S07-hw-dataset-01.csv -> ", X_transformed.shape)

# Использование в S07-hw-dataset-02.csv
X = dataframes["S07-hw-dataset-02.csv"].drop(columns=["sample_id"])
X_transformed = preprocesor["S07-hw-dataset-02.csv"].fit_transform(X)
print("После трансформации  в S07-hw-dataset-02.csv -> ", X_transformed.shape)
```

### Почему исключили категориальные признаки?

В `S07-hw-dataset-04.csv` есть категориальные признаки. Их нужно исключить из обработки для моделей кластеризации. Это связано с тем, что алгоритмы кластеризации такие как `KMeans`, `DBSCAN` и `Agglomerative` из нашего HW работают с числовыми признаками. При преобразование категориальных признаков с помощью `OneHotEncoder` - прилично ухудшит качество кластеризации (для маленьких выборках ну и при наличии шумовых признаков (возможно)), что будет видно из увеличенной в несколько раз размерности.

### **2.3.3. по 2.3.7.**

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

path_artifacts = Path("artifacts/figures/")
path_artifacts.mkdir(parents=True, exist_ok=True)

def kmeans_analysis(X, k_range=range(2, 21), random_state=42, dataset_name=""):
    silhouettes = []
    for k in k_range:
        km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
        labels = km.fit_predict(X)
        score = silhouette_score(X, labels)
        silhouettes.append(score)
    
    # График Silhouette vs KMeans (K)
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, silhouettes, marker='s', linewidth=2, markersize=8, color='steelblue')
    plt.axvline(x=k_range[np.argmax(silhouettes)], color='crimson', linestyle='--', alpha=0.7)
    plt.title(f"Анализ оптимального числа кластеров - {dataset_name}", fontsize=14, fontweight='bold')
    plt.xlabel("Количество кластеров (K)")
    plt.ylabel("Silhouette Score")
    plt.grid(True, alpha=0.3)
    
    fig_name = f"cluster_optimization_analysis_{dataset_name.replace('.csv','').replace(' ', '_')}.png"
    plt.savefig(path_artifacts / fig_name, dpi=200, bbox_inches='tight')
    plt.show()
    
    # Метрики качества для каждого K
    plt.figure(figsize=(12, 6))
    
    # Доп метрики для каждого K
    db_scores = []
    ch_scores = []
    for k in k_range:
        km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
        labels = km.fit_predict(X)
        db_scores.append(davies_bouldin_score(X, labels))
        ch_scores.append(calinski_harabasz_score(X, labels))
    
    # Нормализование метрик для сравнения
    db_norm = np.array(db_scores) / np.max(db_scores)  # Чем меньше, тем лучше
    ch_norm = np.array(ch_scores) / np.max(ch_scores)  # Чем больше, тем лучше
    
    plt.plot(k_range, silhouettes, marker='o', label='Silhouette (↑ лучше)', linewidth=2)
    plt.plot(k_range, 1 - db_norm, marker='s', label='Davies-Bouldin (↑ лучше)', linewidth=2)
    plt.plot(k_range, ch_norm, marker='^', label='Calinski-Harabasz (↑ лучше)', linewidth=2)
    
    plt.title(f"Сравнение метрик кластеризации - {dataset_name}", fontsize=14, fontweight='bold')
    plt.xlabel("Количество кластеров (K)")
    plt.ylabel("Нормализованные значения метрик")
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    fig_name = f"clustering_metrics_comparison_{dataset_name.replace('.csv','').replace(' ', '_')}.png"
    plt.savefig(path_artifacts / fig_name, dpi=200, bbox_inches='tight')
    plt.show()
    
    best_k = k_range[np.argmax(silhouettes)]
    print(f"Лучшее KMeans (K) по silhouette -> {best_k}")
    km_best = KMeans(n_clusters=best_k, n_init=10, random_state=random_state)
    best_labels = km_best.fit_predict(X)
    
    # Размеры кластеров для лучшего K
    plt.figure(figsize=(10, 6))
    cluster_sizes = np.bincount(best_labels)
    bars = plt.bar(range(len(cluster_sizes)), cluster_sizes, color=plt.cm.tab20.colors[:len(cluster_sizes)])
    plt.title(f"Распределение точек по кластерам (K={best_k}) - {dataset_name}", fontsize=14, fontweight='bold')
    plt.xlabel("Номер кластера")
    plt.ylabel("Количество точек")
    
    # Значения на столбцах
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                f'{int(height)}', ha='center', va='bottom', fontsize=10)
    
    plt.grid(True, alpha=0.3, axis='y')
    fig_name = f"cluster_size_distribution_{dataset_name.replace('.csv','').replace(' ', '_')}.png"
    plt.savefig(path_artifacts / fig_name, dpi=200, bbox_inches='tight')
    plt.show()
    
    return best_labels, km_best

def dbscan_analysis(X, eps_list=[0.5, 1, 1.5], min_samples_list=[3, 5, 10], dataset_name=""):
    best_score = -1
    best_params = None
    best_labels = None
    silhouettes_map = []
    
    for eps in eps_list:
        for min_samples in min_samples_list:
            db = DBSCAN(eps=eps, min_samples=min_samples)
            labels = db.fit_predict(X)
            # Кластеров больше 1?
            if len(set(labels)) > 1 and -1 not in set(labels):
                score = silhouette_score(X, labels)
                silhouettes_map.append((eps, min_samples, score))
                if score > best_score:
                    best_score = score
                    best_params = (eps, min_samples)
                    best_labels = labels
    
    # График Silhouette vs eps (при min_samples фикс)
    if silhouettes_map:
        eps_vals, min_samples_vals, scores = zip(*silhouettes_map)
        plt.figure(figsize=(10, 6))
        
        sc = plt.scatter(eps_vals, scores, s=120, 
                        c=min_samples_vals,
                        cmap='viridis', 
                        edgecolor='black', 
                        linewidth=1, 
                        alpha=0.8)
        cbar = plt.colorbar(sc)
        cbar.set_label('min_samples', fontsize=12)
        plt.xlabel('Радиус окрестности (eps)')
        plt.ylabel('Silhouette Score')
        plt.title(f'Параметризация DBSCAN - {dataset_name}', fontsize=14, fontweight='bold')
        plt.grid(True, alpha=0.3)
        
        fig_name = f"dbscan_parameter_tuning_{dataset_name.replace('.csv','').replace(' ', '_')}.png"
        plt.savefig(path_artifacts / fig_name, dpi=200, bbox_inches='tight')
        plt.show()
        
        # Соотношение шума и кластеров в DBSCAN
        if best_labels is not None and -1 in best_labels:
            plt.figure(figsize=(8, 8))
            noise_count = np.sum(best_labels == -1)
            cluster_counts = [np.sum(best_labels == label) for label in set(best_labels) if label != -1]
            
            labels_pie = [f'Кластер {label}' for label in set(best_labels) if label != -1]
            if noise_count > 0:
                labels_pie = ['Шум'] + labels_pie
                sizes = [noise_count] + cluster_counts
                colors_pie = ['gray'] + plt.cm.tab20.colors[1:len(set(best_labels))]
            else:
                sizes = cluster_counts
                colors_pie = plt.cm.tab20.colors[:len(set(best_labels))]
            
            plt.pie(sizes, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%',
            startangle=90, wedgeprops={'edgecolor': 'white', 'linewidth': 2})
            plt.title(f'Соотношение кластеров DBSCAN - {dataset_name}', fontsize=14, fontweight='bold')
            
            fig_name = f"dbscan_cluster_ratio_{dataset_name.replace('.csv','').replace(' ', '_')}.png"
            plt.savefig(path_artifacts / fig_name, dpi=200, bbox_inches='tight')
            plt.show()
    
    if best_params is not None:
        print(f"Лучшие параметры DBSCAN -> eps={best_params[0]}, min_samples={best_params[1]}, silhouette={best_score:.3f}")
    else:
        print("DBSCAN не нашел подходящих кластеров (все точки - шум или только один кластер).")
    return best_labels


for f, preprocessor in preprocesor.items():
    print(f"-----------------Кластеризация, метрики качества и визуализация для {f}-----------------")
    
    X = dataframes[f].drop(columns=["sample_id"])
    X_transformed = preprocessor.fit_transform(X)
    
    # KMeans
    print("\nKMeans")
    k_labels, k_model = kmeans_analysis(X_transformed, dataset_name=f)
    print(f"Silhouette Score -> {silhouette_score(X_transformed, k_labels):.3f}")
    print(f"Davies-Bouldin Score -> {davies_bouldin_score(X_transformed, k_labels):.3f}")
    print(f"Calinski-Harabasz Score -> {calinski_harabasz_score(X_transformed, k_labels):.3f}")
    print("Простая статистика по KMeans ->", np.bincount(k_labels))
    
    # DBSCAN
    print("\nDBSCAN")
    db_labels = dbscan_analysis(X_transformed, dataset_name=f)
    if db_labels is not None:
        noise_mask = (db_labels == -1)
        n_noise = noise_mask.sum()
        total = len(db_labels)
        print(f"Доля шума -> {n_noise / total:.2%} ({n_noise} точек из {total})")
        
        mask = ~noise_mask
        if mask.sum() > 1:
            X_db = X_transformed[mask]
            labels_db = db_labels[mask]
            print(f"Silhouette Score (non-noise) -> {silhouette_score(X_db, labels_db):.3f}")
            print(f"Davies-Bouldin Score (non-noise) -> {davies_bouldin_score(X_db, labels_db):.3f}")
            print(f"Calinski-Harabasz Score (non-noise) -> {calinski_harabasz_score(X_db, labels_db):.3f}")
            print("Простая статистика по DBSCAN  ->", np.bincount(labels_db))
        else:
            print("Нехватает точек без шума для расчёта метрик.")
    else:
        print("DBSCAN не нашел подходящих кластеров.")
    
    print("\nPCA 2D")
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_transformed)
    
    # Наилучший KMeans
    best_labels = k_labels
    
    # Визуализация PCA
    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=best_labels, 
    cmap='viridis', s=70, alpha=0.7, edgecolor='black', linewidth=0.5)
    
    plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
    plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
    plt.title(f'Визуализация кластеризации в 2D пространстве - {f}', fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    fig_name = f"pca_clustering_visualization_{f.replace('.csv','').replace(' ', '_')}.png"
    plt.savefig(path_artifacts / fig_name, dpi=200, bbox_inches='tight')
    plt.show()

print("Готово")
```

### **2.4.** Артефакты эксперимента

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans
import numpy as np

# Действия с S07-hw-dataset-01.csv
fname = "S07-hw-dataset-01.csv"
X = dataframes[fname].drop(columns=["sample_id"])
X_transformed = preprocesor[fname].fit_transform(X)

# Параметры KMeansберу лучшее K сверху
best_k = KMeans(n_clusters=2, n_init=10, random_state=42).fit(X_transformed).n_clusters

# 5 итераций с разными random_state
n_runs = 5
labels_list = []

for i in range(n_runs):
    km = KMeans(n_clusters=best_k, n_init=10, random_state=i*10)
    labels = km.fit_predict(X_transformed)
    labels_list.append(labels)

# Матрицу ARI между всеми парами итераций
ari_matrix = np.zeros((n_runs, n_runs))

for i in range(n_runs):
    for j in range(n_runs):
        ari_matrix[i, j] = adjusted_rand_score(labels_list[i], labels_list[j])

print("Матрица ARI между 5 итерациями KMeans ->", np.round(ari_matrix, 3))

# Средняя устойчивость
upper_tri_indices = np.triu_indices(n_runs, k=1)
mean_ari = ari_matrix[upper_tri_indices].mean()
print(f"\nСреднее ARI между разными итерациями -> {mean_ari:.3f}")

# Интерпретация:
if mean_ari > 0.9:
    print("Разбиения очень стабильны")
elif mean_ari > 0.7:
    print("Разбиения относительно стабильны")
else:
    print("Разбиения нестабильны, KMeans чувствителен к инициализации")
```

```python
import json
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN

def save_metrics_summary(dataframes, preprocessors, metrics_summary_path):
    
    #Собирает и сохраняет метрики для всех датасетов и моделей в json
    summary = {}
    
    for fname, preprocessor in preprocessors.items():
        print(f"Вычисление метрик для {fname}-_-")
        X = dataframes[fname].drop(columns=["sample_id"])
        X_transformed = preprocessor.fit_transform(X)

        # KMeans 
        k_range = range(2, 21)
        silhouettes = []
        for k in k_range:
            km = KMeans(n_clusters=k, n_init=10, random_state=42)
            labels = km.fit_predict(X_transformed)
            score = silhouette_score(X_transformed, labels)
            silhouettes.append(score)
        best_k = k_range[np.argmax(silhouettes)]
        k_labels = KMeans(n_clusters=best_k, n_init=10, random_state=42).fit_predict(X_transformed)

        k_silhouette = silhouette_score(X_transformed, k_labels)
        k_davies_bouldin = davies_bouldin_score(X_transformed, k_labels)
        k_calinski_harabasz = calinski_harabasz_score(X_transformed, k_labels)

        # DBSCAN
        eps_list = [0.5, 1, 1.5]
        min_samples_list = [3, 5, 10]
        best_score = -1
        best_params = None
        best_db_labels = None
        
        for eps in eps_list:
            for min_samples in min_samples_list:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X_transformed)
                if len(set(labels)) > 1 and -1 not in set(labels):
                    score = silhouette_score(X_transformed, labels)
                    if score > best_score:
                        best_score = score
                        best_params = (eps, min_samples)
                        best_db_labels = labels

        if best_db_labels is not None and len(set(best_db_labels)) > 1:
            db_silhouette = silhouette_score(X_transformed, best_db_labels)
            noise_mask = (best_db_labels == -1)
            n_noise = noise_mask.sum()
            total = len(best_db_labels)
            noise_ratio = n_noise / total if total > 0 else 0
            
            mask = ~noise_mask
            if mask.sum() > 1:
                X_db_non_noise = X_transformed[mask]
                labels_db_non_noise = best_db_labels[mask]
                db_davies_bouldin = davies_bouldin_score(X_db_non_noise, labels_db_non_noise)
                db_calinski_harabasz = calinski_harabasz_score(X_db_non_noise, labels_db_non_noise)
            else:
                db_davies_bouldin = None
                db_calinski_harabasz = None
        else:
            db_silhouette = None
            noise_ratio = 1.0
            db_davies_bouldin = None
            db_calinski_harabasz = None
            best_params = None

        summary[fname] = {
            "KMeans": {
                "best_k": int(best_k),
                "silhouette_score": round(k_silhouette, 4),
                "davies_bouldin_score": round(k_davies_bouldin, 4),
                "calinski_harabasz_score": round(k_calinski_harabasz, 4)
            },
            "DBSCAN": {
                "best_params": best_params,
                "silhouette_score": round(db_silhouette, 4) if db_silhouette is not None else None,
                "davies_bouldin_score": round(db_davies_bouldin, 4) if db_davies_bouldin is not None else None,
                "calinski_harabasz_score": round(db_calinski_harabasz, 4) if db_calinski_harabasz is not None else None,
                "noise_ratio": round(noise_ratio, 4)
            }
        }

    os.makedirs(metrics_summary_path.parent, exist_ok=True)
    with open(metrics_summary_path, 'w', encoding='utf-8') as f:
        json.dump(summary, f, indent=4, ensure_ascii=False)
    print(f"Метрики сохранены. Находятся в {metrics_summary_path}.")


def save_best_configs(dataframes, preprocessors, path_configs):

    #Сохраняет инф о лучших параметрах для каждой модели и датасета

    configs = {}
    
    for fname, preprocessor in preprocessors.items():
        print(f"Поиск лучших конфигураций для {fname}-_-")
        X = dataframes[fname].drop(columns=["sample_id"])
        X_transformed = preprocessor.fit_transform(X)

        # KMeans поиск лучшего K
        k_range = range(2, 21)
        silhouettes = []
        for k in k_range:
            km = KMeans(n_clusters=k, n_init=10, random_state=42)
            labels = km.fit_predict(X_transformed)
            score = silhouette_score(X_transformed, labels)
            silhouettes.append(score)
        best_k = k_range[np.argmax(silhouettes)]

        # DBSCAN поиск параметров
        eps_list = [0.5, 1, 1.5]
        min_samples_list = [3, 5, 10]
        best_score = -1
        best_params = None
        for eps in eps_list:
            for min_samples in min_samples_list:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X_transformed)
                if len(set(labels)) > 1 and -1 not in set(labels):
                    score = silhouette_score(X_transformed, labels)
                    if score > best_score:
                        best_score = score
                        best_params = {"eps": eps, "min_samples": min_samples}

        k_silhouette_final = max(silhouettes)
        db_silhouette_final = best_score if best_score != -1 else float('-inf')

        if k_silhouette_final >= db_silhouette_final:
            best_model = "KMeans"
            best_config = {"n_clusters": int(best_k)}
            selection_criterion = "highest_silhouette_score"
        else:
            best_model = "DBSCAN"
            best_config = best_params
            selection_criterion = "highest_silhouette_score"

        configs[fname] = {
            "selected_model": best_model,
            "selected_config": best_config,
            "selection_criterion": selection_criterion,
            "kmeans_info": {
                "best_k_by_silhouette": int(best_k),
                "best_silhouette_score": round(k_silhouette_final, 4)
            },
            "dbscan_info": {
                "best_params_by_silhouette": best_params,
                "best_silhouette_score": round(best_score, 4) if best_score != -1 else None
            }
        }

    os.makedirs(path_configs.parent, exist_ok=True)
    with open(path_configs, 'w', encoding='utf-8') as f:
        json.dump(configs, f, indent=4, ensure_ascii=False)
    print(f"Конфигурации сохранены. Находятся в {path_configs}.")


def save_clust(dataframes, preprocessors, path_label):

    #Сохраняем в csv
    
    os.makedirs(path_label, exist_ok=True)
    
    for fname, preprocessor in preprocessors.items():
        print(f"Сохранение лейблов для {fname}-_-")
        X = dataframes[fname].drop(columns=["sample_id"])
        sample_ids = dataframes[fname]["sample_id"]
        X_transformed = preprocessor.fit_transform(X)

        # Поуск K уже для KMeans
        k_range = range(2, 21)
        silhouettes_k = []
        for k in k_range:
            km = KMeans(n_clusters=k, n_init=10, random_state=42)
            labels = km.fit_predict(X_transformed)
            score = silhouette_score(X_transformed, labels)
            silhouettes_k.append(score)
        best_k = k_range[np.argmax(silhouettes_k)]
        k_silhouette_final = max(silhouettes_k)

        # Поиск параметров уже для DBSCAN
        eps_list = [0.5, 1, 1.5]
        min_samples_list = [3, 5, 10]
        best_score_db = -1
        best_labels_db = None
        for eps in eps_list:
            for min_samples in min_samples_list:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X_transformed)
                if len(set(labels)) > 1 and -1 not in set(labels):
                    score = silhouette_score(X_transformed, labels)
                    if score > best_score_db:
                        best_score_db = score
                        best_labels_db = labels

        # Выбор лучшей модели и генерация лейбл
        if k_silhouette_final >= best_score_db:
            final_labels = KMeans(n_clusters=best_k, n_init=10, random_state=42).fit_predict(X_transformed)
        else:
            if best_labels_db is not None:
                final_labels = best_labels_db
            else:
                print(f"DBSCAN не смог найти кластеры, используем KMeans с K={best_k}.")
                final_labels = KMeans(n_clusters=best_k, n_init=10, random_state=42).fit_predict(X_transformed)

        # Сохранение лейблов
        labels_df = pd.DataFrame({
            'sample_id': sample_ids,
            'cluster_label': final_labels
        })
        ds_num = fname.split("-")[-1].split(".csv")[0]
        output_filename = f"labels_hw07_ds{ds_num}.csv"
        output_path = path_label / output_filename
        labels_df.to_csv(output_path, index=False)
        print(f"Лейблы сохранены. Находятся в {output_path}.")


path_artifacts = Path("artifacts/")
path_metrics = path_artifacts / "metrics_summary.json"
path_configs = path_artifacts / "best_configs.json"
path_label = path_artifacts / "labels/"
path_figures = path_artifacts / "figures/"

save_metrics_summary(dataframes, preprocesor, path_metrics)
save_best_configs(dataframes, preprocesor, path_configs) 
save_clust(dataframes, preprocesor, path_label)

print("Готово")
```
