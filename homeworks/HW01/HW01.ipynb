### **9** Матрица с	0 внутри, и	1 на границах. срезы.

```python
import numpy as np
n = 5
table = np.zeros((n,n))
table[:1] = 1 # первоя строка
table[:,0] = 1 # левый крайний столбец
table[:,n-1] = 1 # правый крайний столбец
table[n-1] = 1 # нижняя строка
print(table)
```
